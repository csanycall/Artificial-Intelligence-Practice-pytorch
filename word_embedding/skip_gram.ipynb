```python
from collections import defaultdict
import torch
import numpy as np
from torch import nn
import random
class WordEmbSkip(nn.Module):
    def __init__(self,nwords,emb_size):
        super(WordEmbSkip,self).__init__()
        self.emb=nn.Embedding(nwords,emb_size)
        self.context=nn.Linear(emb_size,nwords)
    def forward(self,x):
        emb=self.emb(x)
        out=self.context(emb)
        return out
N=2
EMB_SIZE=128
w2i=defaultdict(lambda:len(w2i))
UNK=w2i["<unk>"]
S=w2i["<s>"]

def read_dataset(filename):
    with open(filename,"r") as f:
        for line in f:
            yield[w2i[word] for word in line.lower().strip().split(" ")]
train=list(read_dataset("datalab/39811/train.txt"))
w2i=defaultdict(lambda :UNK,w2i)
dev=list(read_dataset("datalab/39811/valid.txt"))
i2w={v:k for k,v in w2i.items()}
nwords=len(w2i)
```

```python
model=WordEmbSkip(nwords,EMB_SIZE)
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
def sent_loss(sent):
    losses=[]
    for i,word in enumerate(sent):
        for j in range(1, N + 1):
            for direction in [-1,1]:
                c=torch.tensor([word])
                context_id = sent[i + direction * j] if 0 <= i + direction * j < len(sent) else S
                context = torch.tensor([context_id])   # Tensor for context word
                print("111",context.shape)
                logits = model(c)
                loss = criterion(logits, context)
                losses.append(loss)
    return torch.stack(losses).sum()


for epoch in range(100):
    random.shuffle(train)
    train_words,train_loss=0,0.0
    model.train()
    for sent_id,sent in enumerate(train):
        my_loss=sent_loss(sent)
        train_loss+=my_loss.item()
        train_words+=len(sent)
        optimizer.zero_grad()
        my_loss.backward()
        optimizer.step()
        if(sent_id%500==0):
            print("epoch=%r,loss=%.4f"%(epoch,my_loss/train_words))
    print("epoch=%r,loss=%.4f"%(epoch,my_loss/train_words))
    model.eval()
    dev_words,dev_loss=0,0.0
    for sent_id,sent in enumerate(dev):
        my_loss=sent_loss(sent)
        dev_loss+=my_loss.item()
        dev_words+=len(sent)
        if(sent_id%5000==0):
            print("epoch=%r,loss=%.4f"%(epoch,dev_loss/dev_words))
    print("epoch=%r,loss=%.4f"%(epoch,dev_loss/dev_words))
        
```
